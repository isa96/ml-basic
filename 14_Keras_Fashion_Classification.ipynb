```python
import tensorflow as tf

#load the dataset
mnist = tf.keras.datasets.fashion_mnist

#splitting for data train and data test
(x_train, y_train),(x_test, y_test) = mnist.load_data()

#normalization
x_train, x_test = x_train / 255.0, x_test / 255.0

#model dengan 3 layers
model = tf.keras.models.Sequential([
    #layer input dengan pixel gambar 28x28 pixel
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    #hidden layer dengan 512 perseptron dan fungsi aktivasi relu
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    #output layer dengan 10 kelas, maka menggunakan fungsi aktivasi softmax
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

#optimizing dengan Adam optimizer
#karenal >2 kelas maka menggunakan sparse_categorical_crossentropy
#kalau hanya dua kelas gunakan binary_crossentropy
model.compile(optimizer=tf.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

#training model 10 epochs
model.fit(x_train, y_train, epochs=100)
```
